```python
from pydantic import BaseModel, ValidationError
from typing import List, Literal

# (Assuming Unit is defined, adding a minimal version for this example)
class Unit(BaseModel):
    unit_number: int
    unit_name: str
    topics: List[str]

# 1. Define your specific Literal type with all valid department strings
ValidDepartments = Literal[
    'cse', 'it', 'electronics', 'electrical', 'mechanical', 
    'meta', 'civil', 'chemical', 'biotech', 'biomed', 'mining'
]

# 2. Use that type in your model
class Subject(BaseModel):
    name: str
    semester: int
    department: ValidDepartments  # <-- This is the fix
    credits: str
    status: str
    code: str
    pre_requisites: List[str]
    units: List[Unit]
    course_material: List[str]
```

```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
```

```python
docs = []
from pathlib import Path
for el in syllabus_links:
    file_path = Path(f"{syllabus_dir/el[0]}.pdf")
    
    if not file_path.is_file():
        continue
        
    cur_docs = PyPDFLoader(file_path).load()[1:]
    cur_res_docs = []
    for cur_doc in cur_docs:
        if cur_doc.page_content:
            cur_res_docs.append(cur_doc)

    docs.extend(cur_res_docs)
    print(f"loaded {len(cur_res_docs)} out of {len(cur_docs)} from {el[0]}")
```

```python
system_prompt = """
You are an expert, high-accuracy document parsing assistant.
Your task is to read the provided syllabus document text and extract detailed information for all subjects found within it.

Your output MUST be a single, valid JSON list `[...]`, where each item in the list is a JSON object representing one `Subject`.

---
### 1. Critical Parsing Heuristic (How to find a valid subject)

You MUST use the following rules to identify a valid, parsable subject page. If a page or document does not strictly adhere to this format, **it must be skipped.**

1.  **Check for Subject Name:** A valid subject page MUST begin with the subject's name (e.g., "Data Warehousing and Data Mining").
2.  **Skip Invalid Pages:** Any page that *does not* start with a clear subject name, or that appears to be an introductory table of contents (like Page 1 of the example, which lists multiple subjects), **must be ignored.**
3.  **Verify Required Sections:** A valid subject page MUST contain the following key sections, typically in this order:
    * Subject Name
    * Semester (e.g., "[6th Semester, Third Year]")
    * Credits, Status, and Code
    * Course Content (which is broken down into "Unit-1", "Unit-2", etc.)
    * Each unit consist of ',' seperated topics. They must be split into a list.
    * Course Materials (with "Required Text" and/or "Optional Materials")
4.  department should be one of the following: [cse, it, electronics, electrical, mechanical, meta, civil, chemical, mining, biotech, biomed]
"""
```

```python
import asyncio
from langchain_google_genai import ChatGoogleGenerativeAI
from tqdm import tqdm

# Create multiple LLM clients (helps spread rate limits)
def make_llm():
    return ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        temperature=0
    ).with_structured_output(Subject)

# Split docs into batches of size 10
batch_size = 10
batches = [docs[i:i+batch_size] for i in range(0, len(docs), batch_size)]

async def process_batch(batch, llm):
    return await llm.abatch([
        [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": doc.page_content}
        ]
        for doc in batch
    ])

async def main():
    all_subjects = []

    # Create a small pool of LLMs for concurrency
    llms = [make_llm() for _ in range(5)]  # parallel workers

    # Process batches concurrently
    tasks = [
        process_batch(batches[i], llms[i % len(llms)])
        for i in range(len(batches))
    ]

    for f in tqdm(asyncio.as_completed(tasks), total=len(tasks)):
        result = await f
        all_subjects.extend(result)

    return all_subjects

# Run
all_subjects = await main()
```

```python
import sqlite3
```

```python
conn = sqlite3.connect(root_dir / "data" / "syllabus.db")
cur = conn.cursor()
```

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS subjects (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    semester INTEGER,
    department TEXT,
    credits TEXT,
    status TEXT,
    code TEXT,
    pre_requisites TEXT,   -- JSON
    units TEXT,            -- JSON
    course_material TEXT   -- JSON
)
""")

conn.commit()
```

```python
def validate_subject(sub: Subject) -> bool:
    if sub.semester == 0:
        return False
    if not sub.name:
        return False
    if not sub.department:
        return False
    if not sub.credits:
        return False
    if len(sub.units) == 0:
        return False
    return True
```

```python
import json
data = [
    (
        s.name,
        s.semester,
        s.department,
        s.credits,
        s.status,
        s.code,
        json.dumps(s.pre_requisites),
        json.dumps([u.model_dump() for u in s.units]),
        json.dumps(s.course_material)
    )
    for s in all_subjects if s and validate_subject(s)
]
```

```python
cur.execute("SELECT * FROM subjects")
rows = cur.fetchall()
```

```python
columns = [desc[0] for desc in cur.description]
print("Columns:", columns)
```

```python
subjects_loaded = []
for row in rows:
    _, name, semester, department, credits, status, code, pre, units, material = row

    subject = Subject(
        name=name,
        semester=semester,
        department=department,
        credits=credits,
        status=status,
        code=code,
        pre_requisites=json.loads(pre),
        units=[Unit(**u) for u in json.loads(units)],
        course_material=json.loads(material)
    )

    subjects_loaded.append(subject)
len(subjects_loaded)
```
